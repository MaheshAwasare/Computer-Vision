```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sachin1 = cv2.imread('images/Sachin1.jpg',0)
```

```python
plt.imshow(sachin1)
```

```python
plt.imshow(sachin1,cmap='gray')
```

```python
f_cascade = cv2.CascadeClassifier('cascade_data\haarcascade_frontalcatface.xml')
```

```python
def detect_face(img):
    output_img = img.copy()
    face_detector = f_cascade.detectMultiScale(output_img,scaleFactor=1.2, minNeighbors=1)
    
    for(x,y,w,h) in face_detector:
        cv2.rectangle(output_img,(x,y),(x+w, y+h),(205,92,92),5)
    
    return output_img
```

```python
output = detect_face(sachin1)
```

```python
plt.imshow(output)
```

<h2>Video Detector </h2>

```python
myvid = cv2.VideoCapture(0)

while True:
    ret,frame = myvid.read(0)
    frame = detect_face(frame)
    cv2.imshow('Detecting Mahesh!!!!', frame)
    waitKey = cv2.waitKey(1)
    if waitKey == 27:
        break
myvid.release()
cv2.destroyAllWindows()
```
